````python
students_data = {"student_id" : [1, 2, 3, 4, 5, 6],
                "name" : ['Ram', 'Sita', 'Hari', 'Gita', 'Suman', 
                          'Ravi'],
                "age" : [21, 22, 20, 23, 24, 22],
                "city" : ['Kathmandu', 'Pokhara', 'Kathmandu', 'Butwal', 
                          'Pokhara', 'Kathmandu'],
                "marks_math" : [80, 92, 75, 68, 60, 85],
                "marks_english" : [78, 88, 72, 70, 65, 90]}
````

````python
import pandas as pd
````

````python
df_students = pd.DataFrame(students_data)
df_students
````

## What range are the marks of math in?

````python
sales_data = {
    "sales_id" : [101, 102, 103, 104, 105, 106, 107, 108],
    "product" : ["A", "A", "B", "B", "C", "A", "C", "B"],
    "region" : ["East", "West", "East", "West", "East", "East", "West", 
                "East"],
    "units_sold" : [10, 15, 7, 20, 5, 12, 8, 9],
    "price_per_unit" : [100, 100, 150, 150, 200, 100, 200, 150]
}

df_sales = pd.DataFrame(sales_data)
df_sales
````

````python
df_sales['revenue'] = df_sales['units_sold'] * df_sales['price_per_unit']
````

````python
df_sales
````
