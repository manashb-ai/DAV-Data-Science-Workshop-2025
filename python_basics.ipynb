## Create a dictionary for 3 students. Each student should have a name and a single total_marks value. Find and print the name of the student with the highest marks.

````python
students = {"Ram" : 80,
           "Shyam" : 90,
           "Ramesh": 93,
           "Kumar": 56}

highest_score = 0
highest_score_name = ""
for name in students:
    if students[name] > highest_score:
        highest_score = students[name]
        highest_score_name = name

print(f"The student with the highest score of {highest_score} is {highest_score_name}.")
````

## Write a function `merge_dicts(d1, d2)` that merges two dictionaries and returns the result.

````python
d1 = {"a" : 1, "b" : 2}
d2 = {"c" : 3, "d" : 4}

def merge_dicts(d1, d2):
    for k in d1:
        d2[k] = d1[k]
    return d2

merge_dicts(d1, d2)
````

# Python Lists vs NumPy

````python
import time
import numpy as np

n = 10000000

py_list1 = list(range(n))
py_list2 = list(range(n))

start = time.perf_counter()
py_result = [a * b for a, b in zip(py_list1, py_list2)]
end = time.perf_counter()
print(f"Python list time: {end - start:.5f} seconds")

np_array1 = np.arange(n)
np_array2 = np.arange(n)

start = time.perf_counter()
np_result = np_array1 * np_array2
end = time.perf_counter()
print(f"NumPy array time: {end - start:.5f} seconds")
````
